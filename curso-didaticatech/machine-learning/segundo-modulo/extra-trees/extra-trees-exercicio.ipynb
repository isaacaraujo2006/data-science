```python
#Separando as variáveis entre preditoras e variável target
y = arquivo['mushroom']
x = arquivo.drop('mushroom', axis=1)
```

```python
#Transformando variáveis categóricas em numeros (one hot encode)
x_encoding = pd.get_dummies(x)
```

```python
x_encoding.head()
```

```python
pd.set_option('display.max_rows', 117)
faltantes = x_encoding.isnull().sum()
faltantes_percentual = (x_encoding.isnull().sum() / len(y)) * 100
print(faltantes_percentual)
```

```python
y = y.replace('EDIBLE', 0)
y = y.replace('POISONOUS', 1)
```

```python
y.value_counts()
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score

#Criação do modelo:
modelo = ExtraTreesClassifier(n_estimators=100, n_jobs=-1)#default 10 árvores de decisão.
skfold = StratifiedKFold(n_splits=3, shuffle=True)
resultado = cross_val_score(modelo,x_encoding,y,cv = skfold, n_jobs=-1)
print(resultado.mean())
```
